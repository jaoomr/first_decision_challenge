```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
```

```python
#Como o algoritmo de TF-IDF não encontrou com muita eficiência muitas empresas
#E o mais eficiente não quis utilizar muitas cotas de request
#Farei uma EDA do dataset como um todo junto
water_companies = pd.read_csv("water_companies_by_similarities.csv")
companies = pd.read_csv("data.csv")
```

```python
most_common_cities_water = water_companies['city'].value_counts().head(5)
print("Cidades com mais empresas de água:\n", most_common_cities_water)
```

```python
most_common_cities = companies['city'].value_counts().head(5)
print("Cidades com mais empresas:\n", most_common_cities)
```

```python
water_biggest_companies = water_companies.nlargest(5, 'employees')
print("\Maiores empresas de tratamento de água:\n", water_biggest_companies[['name', 'employees']])
```

```python
companies['employees'] = pd.to_numeric(companies['employees'], errors='coerce')
biggest_companies = companies.nlargest(5, 'employees')
print("\nMaiores empresas do canadá:\n", biggest_companies[['name', 'employees']])
```

```python
#Setando o mapa do canadá para plotar as empresas de água
canada_map_water_companies = folium.Map(location=[56.1304, -106.3468], zoom_start=4)
```

```python
#Plotando o mapa do canadá com pontos de localização das companhias de água
for idx, row in water_companies.iterrows():
    folium.Marker([row['lat'], row['lng']], popup=row['name']).add_to(canada_map_water_companies)
```

```python
#Setando o mapa do canadá para plotar todas empresas
#Preferi usar heatmap para não poul
#Limparei o dataset sem as informações para poder fazer o plot

companies['lat'] = pd.to_numeric(companies['lat'], errors='coerce')
companies['lng'] = pd.to_numeric(companies['lng'], errors='coerce')
companies.dropna(subset=['lat','lng','subcountry'], inplace=True)


canada_heatmap = folium.Map(location=[56.1304, -106.3468], zoom_start=4)
for subcountry, sub_df in companies.groupby('subcountry'):
    heat_data = [[row['lat'], row['lng']] for index, row in sub_df.iterrows()]
    HeatMap(heat_data).add_to(canada_heatmap)
```
